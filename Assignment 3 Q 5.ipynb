```python
import numpy as np
import time
import matplotlib.pyplot as plt

def direct_dft(x):
    N = len(x)
    X = np.zeros(N, dtype=np.complex128)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)
    return X

# Function to measure time taken by direct DFT computation
def measure_direct_dft_time(n_values):
    times_direct = []
    for n in n_values:
        x = np.random.rand(n) + 1j * np.random.rand(n)  # Generating complex numbers for input
        start_time = time.time()
        direct_dft(x)
        end_time = time.time()
        times_direct.append(end_time - start_time)
    return times_direct

# Function to measure time taken by numpy's FFT
def measure_fft_time(n_values):
    times_fft = []
    for n in n_values:
        x = np.random.rand(n) + 1j * np.random.rand(n)  # Generating complex numbers for input
        start_time = time.time()
        np.fft.fft(x)
        end_time = time.time()
        times_fft.append(end_time - start_time)
    return times_fft

# Define the range of n values
n_values = range(4, 101)

# Measure time taken by both methods
times_direct = measure_direct_dft_time(n_values)
times_fft = measure_fft_time(n_values)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(n_values, times_direct, label='Direct DFT', marker='o')
plt.plot(n_values, times_fft, label='NumPy FFT', marker='o')
plt.title('Time Taken by Direct DFT vs NumPy FFT')
plt.xlabel('Number of Samples (n)')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```
