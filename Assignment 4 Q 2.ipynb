```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(10000)

plt.hist(data, bins=50, density=True, alpha=0.6, color='b', label='Histogram')

x_vals = np.linspace(0, 1, 100)
y_vals = np.ones_like(x_vals)
plt.plot(x_vals, y_vals, 'r-', lw=2, label='Uniform PDF')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram vs Uniform PDF')
plt.legend()

plt.show()
```
