```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions g(x) and h(x)
def g(x):
    return np.exp(-x**2)

def h(x):
    return np.exp(-4*x**2)

# Define the range of x values
x = np.linspace(-5, 5, 1000)
dx = x[1] - x[0]

# Compute the functions g(x) and h(x)
g_x = g(x)
h_x = h(x)

# Compute the Fourier transforms of g(x) and h(x)
G_k = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(g_x))) * dx
H_k = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(h_x))) * dx

# Compute the product of the Fourier transforms
convolution_k = G_k * H_k

# Compute the inverse Fourier transform to obtain the convolution
convolution = np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(convolution_k))) / dx

# Plotting the results
plt.figure(figsize=(10, 6))

plt.plot(x, convolution.real, label='Convolution')
plt.plot(x, g_x, label='g(x)')
plt.plot(x, h_x, label='h(x)')
plt.title('Convolution of g(x) and h(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
