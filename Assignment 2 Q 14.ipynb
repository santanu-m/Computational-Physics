```python
import numpy as np
import matplotlib.pyplot as plt

def euler_system(f, t0, y0, h, n):
    t = np.linspace(t0, t0 + n * h, n + 1)
    y = np.zeros((n + 1, len(y0)))
    y[0] = y0
    for i in range(n):
        y[i + 1] = y[i] + h * f(t[i], y[i])
    return t, y

def system(t, y):
    y1, y2 = y
    dy1_dt = y2
    dy2_dt = (2 * t * y2 - 2 * y1 + t**3 * np.log(t)) / t**2
    return np.array([dy1_dt, dy2_dt])

t0, y0 = 1, np.array([1, 0])
h = 0.001
n = 1000

t, y = euler_system(system, t0, y0, h, n)
# plt.figure(figsize=(20, 15))
plt.plot(t, y[:, 0], label='y(t)',marker='o', markersize=2)
plt.plot(t,(7/4) * t + ((t**3) / 2) * np.log(t) - (3/4) * (t**3),label="Analytic solution")
plt.title('Solution of ' r"$t^2 y'' - 2ty'+2y=t^3$")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.grid(True)
plt.show()
```
