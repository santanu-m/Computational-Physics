```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import sympy as sp

def ode(t, y):
    return t * np.exp(3 * t) - 2 * y

t0, tf = 0, 1
y0 = [0]

sol = solve_ivp(ode, [t0, tf], y0, t_eval=np.linspace(t0, tf, 100))

t, C = sp.symbols('t C')
y = sp.Function('y')(t)
expr = sp.Eq(sp.diff(y, t), t * sp.exp(3*t) - 2*y)
sol_symbolic = sp.dsolve(expr, y, ics={y.subs(t, 0): 0})

y_exact = sp.lambdify(t, sol_symbolic.rhs, 'numpy')

plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], 'r-', label='Numerical solution')
plt.plot(sol.t, y_exact(sol.t), 'b--', label='Analytical solution')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title("Solution of " r"$t \ e^{3t}-2y$")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def ode(t, y):
    return 1 - (t - y)**2

# Solve the IVP numerically
t0, tf = 2, 3
y0 = [1]
solution = solve_ivp(ode, [t0, tf], y0, t_eval=np.linspace(t0, tf, 100))

# Analytical solution
t, C = sp.symbols('t C')
y = sp.Function('y')(t)

# Equation for analytical solution using sympy
expr = sp.Eq(sp.diff(y, t), 1 - (t - y)**2)
sol_symbolic = sp.dsolve(expr, y, ics={y.subs(t, 2): 1})

# Convert symbolic solution to a numerical function
y_exact = sp.lambdify(t, sol_symbolic.rhs, 'numpy')

# Evaluate the analytical solution over the same range
t_vals = np.linspace(t0, tf, 100)
y_vals = y_exact(t_vals)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(solution.t, solution.y[0], 'r-', label='Numerical solution')
plt.plot(t_vals, y_vals, 'b--', label='Analytical solution')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Numerical and Analytical Solutions of ' r'$1-(t-y)^2$')
plt.legend()
plt.grid(True)
plt.show()
```
