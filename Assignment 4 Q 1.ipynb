```python
import numpy as np
import matplotlib.pyplot as plt

class LinearCongruentialGenerator:
    def __init__(self, seed, a=1664525, c=1013904223, m=2**32):
        self.seed = seed
        self.a = a
        self.c = c
        self.m = m
        self.current = seed
    
    def next(self):
        self.current = (self.a * self.current + self.c) % self.m
        return self.current / self.m

lcg = LinearCongruentialGenerator(seed=42)
random_numbers = [lcg.next() for _ in range(10000)]

plt.hist(random_numbers, bins=50, density=True, alpha=0.6, color='g', label='LCG Histogram')

x = np.linspace(0, 1, 100)
uniform_pdf = np.ones_like(x)
plt.plot(x, uniform_pdf, 'r-', lw=2, label='Uniform PDF')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of LCG Random Numbers vs Uniform PDF')
plt.legend()
plt.show()
```
