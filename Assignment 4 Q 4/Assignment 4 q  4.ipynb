```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

file_path = 'Downloads/exponential_data.txt'
data = np.loadtxt(file_path)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Data Histogram')

loc, scale = expon.fit(data, floc=0)  # force the location to 0
x = np.linspace(0, np.max(data), 1000)
pdf_fitted = expon.pdf(x, loc, scale)
plt.plot(x, pdf_fitted, 'r-', label='Exponential PDF')

plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Histogram of Data and Fitted Exponential PDF')
plt.legend()
plt.show()
```
