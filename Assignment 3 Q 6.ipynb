```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constant function
def constant_function(x, c):
    return c * np.ones_like(x)

# Define the range of x values
x = np.linspace(-10, 10, 1000)

# Define the constant value
c = 2

# Compute the constant function values
f_x = constant_function(x, c)

# Compute the Fourier transform using NumPy's FFT
f_k = np.fft.fftshift(np.fft.fft(np.fft.ifftshift(f_x))) / (x[1] - x[0])

# Define the k values corresponding to the Fourier transform
k = 2 * np.pi * np.fft.fftshift(np.fft.fftfreq(len(x), d=(x[1] - x[0])))

# Plotting the results
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x, f_x, label='Constant Function')
plt.title('Constant Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(k, np.real(f_k), label='Real part')
plt.plot(k, np.imag(f_k), label='Imaginary part')
plt.title('Fourier Transform')
plt.xlabel('Frequency (k)')
plt.ylabel('F(k)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
