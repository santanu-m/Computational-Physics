```python
import numpy as np
import matplotlib.pyplot as plt

def metropolis_sampling(n_samples, initial_state, proposal_std):
    samples = []
    current_state = initial_state
    
    for _ in range(n_samples):
        proposed_state = current_state + np.random.normal(0, proposal_std)
        
        if 3 < proposed_state < 7:
            current_state = proposed_state
        
        samples.append(current_state)
    
    return samples

n_samples = 10000
initial_state = 5  
proposal_std = 0.5 
samples = metropolis_sampling(n_samples, initial_state, proposal_std)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(samples, lw=0.5)
plt.title("Markov Chain")
plt.xlabel("Iteration")
plt.ylabel("X")

plt.subplot(1, 2, 2)
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g', label='MCMC Histogram')

x = np.linspace(3, 7, 1000)
target_density = np.ones_like(x) / (7 - 3)
plt.plot(x, target_density, 'r-', lw=2, label='Target Distribution (Uniform)')

plt.xlabel('x')
plt.ylabel('Density f(x)')
plt.title('Density Histogram of Samples')
plt.legend()
plt.tight_layout()
plt.show()
```
