```python
import numpy as np
import matplotlib.pyplot as plt

def box_muller_transform(n):
    """ Generate n samples using the Box-Muller transform """
    u1 = np.random.uniform(0, 1, n)
    u2 = np.random.uniform(0, 1, n)
    r = np.sqrt(-2 * np.log(u1))
    theta = 2 * np.pi * u2
    z0 = r * np.cos(theta)
    z1 = r * np.sin(theta)
    return z0, z1

n_samples = 10000
z0, z1 = box_muller_transform(n_samples // 2)
gaussian_samples = np.concatenate((z0, z1))

plt.hist(gaussian_samples, bins=50, density=True, alpha=0.6, color='g', label='Histogram')
x = np.linspace(-4, 4, 1000)
pdf = (1/np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)
plt.plot(x, pdf, 'r-', lw=2, label='Gaussian PDF')

plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Histogram of Gaussian Distribution with Mean 0 and Variance 1')
plt.legend()
plt.show()
```
