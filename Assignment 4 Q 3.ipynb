```python
import numpy as np
import matplotlib.pyplot as plt
import time

def lcg(n, seed=0):
    a = 1664525
    c = 1013904223
    m = 2**32
    X = seed
    numbers = []
    for _ in range(n):
        X = (a * X + c) % m
        numbers.append(X / m)
    return np.array(numbers)

n = 10000

start_time = time.time()
lcg_numbers = lcg(n)
lcg_time = time.time() - start_time

start_time = time.time()
np_random_numbers = np.random.rand(n)
np_time = time.time() - start_time

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(lcg_numbers, bins=50, density=True, alpha=0.75, label='LCG Histogram')
x = np.linspace(0, 1, 100)
plt.plot(x, np.ones_like(x), 'r-', label='Uniform PDF')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('LCG Random Numbers')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(np_random_numbers, bins=50, density=True, alpha=0.75, label='np.random.rand Histogram')
plt.plot(x, np.ones_like(x), 'r-', label='Uniform PDF')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('np.random.rand Random Numbers')
plt.legend()

plt.tight_layout()
plt.show()

print(f"LCG generation time: {lcg_time:.6f} seconds")
print(f"np.random.rand generation time: {np_time:.6f} seconds")
```
